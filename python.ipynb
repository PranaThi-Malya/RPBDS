```python
import pandas as pd
```

```python
df=pd.read_csv("/content/data_ted_talks.csv")
```

```python
df
```

```python
# prompt: linear regression graph

import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Select the features and target variable
X = df[['views']]  # Features (independent variable)
y = df['duration']  # Target variable (dependent variable)

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)

# Plot the data and the regression line
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red')
plt.xlabel('Views')
plt.ylabel('duration')
plt.title('Linear Regression of Views vs duration')
plt.show()
```
